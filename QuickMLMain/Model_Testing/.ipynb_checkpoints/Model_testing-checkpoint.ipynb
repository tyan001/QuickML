```python
import tensorflow as tf
import json
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = tf.keras.models.load_model('rps.h5')
```

```python
with open('rps_Label_Names.json', 'r') as f:
    labels = json.load(f)
```

```python
img = tf.keras.preprocessing.image.load_img('rps/testing/papers/paper01-021.png',
                                           target_size=(150,150))
```

```python
plt.imshow(img)
```

```python
img = tf.keras.preprocessing.image.img_to_array(img)
```

```python
pred = model.predict(np.expand_dims(img, axis=0))
```

```python
get_label(np.argmax(pred), labels)
```
